```python
from sklearn.model_selection import train_test_split,GridSearchCV,RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import recall_score,confusion_matrix

from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.decomposition import PCA
```

```python
x=df.drop(columns=['Heart_Risk'])
y=df['Heart_Risk']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
paramters={
    'Classifier__criterion':['gini','entropy','log_loss'],
    'Classifier__n_estimators':[80,100],
    'Classifier__max_depth':[1,2,3,4,5,6,7], 
    'Classifier__min_samples_split':[1,2,3,4,5,6,7], 
    'Classifier__min_samples_leaf':[1,2,3,4,5,6,7], 
    'Classifier__max_features':['auto','sqrt','log2']
}
```

```python
pipe=Pipeline(steps=[
    ('Scaling',StandardScaler()),
    ('PCA',PCA(n_components=3)),
    ('Classifier',RandomForestClassifier(random_state=42))
])
```

```python
#cv=GridSearchCV(pipe,param_grid=paramters, cv=3, n_jobs=-1,scoring='recall')
cv= RandomizedSearchCV(
    estimator=pipe,
    param_distributions=paramters,
    n_iter=10,
    scoring='recall',
    cv=5,
    random_state=42,
    n_jobs=-1
)
```

```python
cv.fit(x_train,y_train)
```

```python
cv.best_params_
```

```python
y_pred=cv.predict(x_test)
```

```python
recall_score(y_test,y_pred)
```

```python
confusion_matrix(y_test,y_pred)
# Here Precision and Recall are high which is good(TP,TN)
```

```python
print("Train Recall:", recall_score(y_train, cv.predict(x_train)))
print("Test Recall:", recall_score(y_test, cv.predict(x_test)))
```
