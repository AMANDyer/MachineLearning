```python
x=df.drop(columns=["Attrition"])
y=df["Attrition"]
```

```python
lb=LabelEncoder()
```

```python
y_encode=lb.fit_transform(y)
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y_encode,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
numeric_colms=x_train.select_dtypes(include='int').columns.to_list()
categorical_colms=x_train.select_dtypes(include='object').columns.to_list()
```

```python
categorical_colms
```

```python
sc=StandardScaler()
```

```python
numeric_transform=Pipeline(steps=[
    ('Scaling',StandardScaler())
])
cat_transform=Pipeline(steps=[
    ('Encode',OneHotEncoder(drop='first',sparse_output=False,handle_unknown='ignore'))
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('trf1',numeric_transform,numeric_colms),
    ('trf2',cat_transform,categorical_colms)
],remainder='passthrough')
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
import xgboost
from xgboost import XGBClassifier

from sklearn.metrics import accuracy_score
```

```python
#Logistic_regression
model_1=Pipeline(steps=[('trf1',preprocessor),
                       ('trf2',LogisticRegression())])
```

```python
model_1.fit(x_train,y_train)
```

```python
y_pred1=model_1.predict(x_test)
```

```python
accuracy_score(y_test,y_pred1)
```

```python
#DT_classification
model_2=Pipeline(steps=[('trf1',preprocessor),
                       ('trf2',DecisionTreeClassifier())])
```

```python
model_2.fit(x_train,y_train)
```

```python
y_pred2=model_2.predict(x_test)
```

```python
accuracy_score(y_test,y_pred2)
```

```python
#RF_classification
model_3=Pipeline(steps=[('trf1',preprocessor),
                       ('trf2',RandomForestClassifier())])
```

```python
model_3.fit(x_train,y_train)
```

```python
y_pred3=model_3.predict(x_test)
```

```python
accuracy_score(y_test,y_pred3)
```

```python
#SVC_classification
model_4=Pipeline(steps=[('trf1',preprocessor),
                       ('trf2',SVC())])
```

```python
model_4.fit(x_train,y_train)
```

```python
y_pred4=model_4.predict(x_test)
```

```python
accuracy_score(y_test,y_pred4)
```

```python
#knn_classification
model_5=Pipeline(steps=[('trf1',preprocessor),
                       ('trf2',KNeighborsClassifier())])
```

```python
model_5.fit(x_train,y_train)
```

```python
y_pred5=model_5.predict(x_test)
```

```python
accuracy_score(y_test,y_pred5)
```

```python
#XGB_classification
model_6=Pipeline(steps=[('trf1',preprocessor),
                       ('trf2',XGBClassifier())])
```
