```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("Attrition - Attrition.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.drop(columns=["Attrition"])
y=df["Attrition"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder,LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
lb=LabelEncoder()
```

```python
y=lb.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.head()
```

```python
y_train
```

```python
categorical_features=x_train.select_dtypes(include="object").columns
```

```python
categorical_features=categorical_features.tolist()
```

```python
categorical_features
```

```python
numerical_columns=x_train.select_dtypes(include="int").columns.tolist()
```

```python
numerical_columns
```

```python
numeric_transform=Pipeline(steps=[
    ('Scaling',StandardScaler())
])
categorical_transform=Pipeline(steps=[
    ('Encoding',OneHotEncoder(drop='first',sparse_output=False,handle_unknown='ignore'))
])
```

```python
preprocessor=ColumnTransformer(transformers=[
    ('trf1',numeric_transform,numerical_columns),
    ('trf2',categorical_transform,categorical_features)
],remainder='passthrough')
```

Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
pipe=Pipeline(steps=[
    ('prerocess',preprocessor),
    ('Linear_model',LogisticRegression(penalty='elasticnet',solver='saga',l1_ratio=0.5,
    C=1.0))
])
```

```python
y_train
```

```python
pipe.fit(x_train,y_train)
```

```python
y_pred=pipe.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

DecisionTree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
pipe_DT=Pipeline(steps=[
    ('prerocess',preprocessor),
    ('DecisionT_model',DecisionTreeClassifier())
])
```

```python
pipe_DT.fit(x_train,y_train)
```

```python
y_pred=pipe_DT.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

SVM

```python
from sklearn.svm import SVC
```

```python
pipe_SVM=Pipeline(steps=[
    ('preprocess',preprocessor),
    ('SVM_model',SVC(kernel='linear'))
])
```

```python
pipe_SVM.fit(x_train,y_train)
```

```python
y_pred=pipe_SVM.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

RandomForest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
pipe_RFC=Pipeline(steps=[
    ('preprocess',preprocessor),
    ('RFC_model',RandomForestClassifier())
])
```

```python
pipe_RFC.fit(x_train,y_train)
```

```python
y_pred=pipe_RFC.predict(x_test)
```

XGBoost Classifier

```python
from xgboost import XGBClassifier
```

```python
pipe_XGC=Pipeline(steps=[
    ('preprocess',preprocessor),
    ('XGC_model',XGBClassifier(n_estimators=100,     # number of boosting rounds
    max_depth=3,          # depth of each tree
    learning_rate=0.1,    # step size shrinkage
    subsample=0.8,        # fraction of samples used per tree
    colsample_bytree=0.8, # fraction of features used per tree
    random_state=42
))
])
```

```python
pipe_XGC.fit(x_train,y_train)
```

```python
y_pred=pipe_XGC.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```

```python
y_train_pred=pipe_XGC.predict(x_train)
```

```python
accuracy_score(y_train,y_train_pred)
```
